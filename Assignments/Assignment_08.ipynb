## Home work Question 1:
Following what we did in class, further break down the category of poor countries and use additional statuses:

- Poor and somewhat healthy
- Poor and somewhat unhealthy

Then make the choropleth plot.

```python
# Rename columns for simplicity
df = df.rename(columns={
    "GDP per capita (current US$)": "GDP per Capita",
    "Life expectancy at birth, total (years)": "Life Expectancy",
    "Population, total": "Population"
})

# Filter for year 2020 and drop rows with missing GDP or Life Expectancy
df_2020 = df[df["Year"] == 2020].dropna(subset=["GDP per Capita", "Life Expectancy"])

# Calculate quartiles
gdp_quartiles = df_2020['GDP per Capita'].quantile([0.25, 0.5, 0.75])
life_exp_quartiles = df_2020['Life Expectancy'].quantile([0.25, 0.5, 0.75])

# Define the functions to categorize GDP and Life Expectancy into quartiles
def assign_gdp_quartile(gdp):
    if gdp <= gdp_quartiles[0.25]:
        return "1st Quartile (<=25%)"
    elif gdp <= gdp_quartiles[0.5]:
        return "2nd Quartile (25-50%)"
    elif gdp <= gdp_quartiles[0.75]:
        return "3rd Quartile (50-75%)"
    else:
        return "4th Quartile (>75%)"

def assign_life_exp_quartile(life_exp):
    if life_exp <= life_exp_quartiles[0.25]:
        return "1st Quartile (<=25%)"
    elif life_exp <= life_exp_quartiles[0.5]:
        return "2nd Quartile (25-50%)"
    elif life_exp <= life_exp_quartiles[0.75]:
        return "3rd Quartile (50-75%)"
    else:
        return "4th Quartile (>75%)"

# Apply quartile functions to the dataframe
df_2020["GDP Quartile"] = df_2020["GDP per Capita"].apply(assign_gdp_quartile)
df_2020["Life Expectancy Quartile"] = df_2020["Life Expectancy"].apply(assign_life_exp_quartile)

# Extend health and wealth classification
def find_status(row):
    if row["GDP Quartile"] == "4th Quartile (>75%)":
        if row["Life Expectancy Quartile"] == "1st Quartile (<=25%)":
            return "Poor and unhealthy"
        elif row["Life Expectancy Quartile"] == "2nd Quartile (25-50%)":
            return "Poor and somewhat unhealthy"
        elif row["Life Expectancy Quartile"] == "3rd Quartile (50-75%)":
            return "Poor and somewhat healthy"
        else:
            return "Poor but healthy"
    elif row["GDP Quartile"] == "1st Quartile (<=25%)" and row["Life Expectancy Quartile"] == "4th Quartile (>75%)":
        return "Rich and healthy"
    elif row["GDP Quartile"] == "4th Quartile (>75%)" and row["Life Expectancy Quartile"] == "1st Quartile (<=25%)":
        return "Rich but unhealthy"
    else:
        return "Other"

df_2020["Health and Wealth Status"] = df_2020.apply(find_status, axis=1)

# Plot the choropleth map
fig = px.choropleth(
    df_2020,
    locations="Country",
    locationmode="country names",
    color="Health and Wealth Status",
    hover_name="Country",
    title="Global Health and Wealth Status in 2020"
)

fig.show()
```

**1. Use Functions for Repeated Logic**:

Encapsulate repetitive logic (such as assigning quartiles) in functions. This reduces the need to rewrite code and minimizes the chances of errors. For example, use a single function to handle quartile assignments for both GDP and life expectancy.


```python
def assign_quartile(value, quartiles, labels):
    """Assigns a quartile based on predefined boundaries and labels."""
    if value <= quartiles[0.25]:
        return labels[0]
    elif value <= quartiles[0.5]:
        return labels[1]
    elif value <= quartiles[0.75]:
        return labels[2]
    else:
        return labels[3]
```
