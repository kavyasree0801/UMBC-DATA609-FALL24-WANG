```python
df3.shape
```

## Homework Question 1:

to check if this dataset has duplicate rows. If it has, only keep one from the duplicated rows and drop the rest.

```python
# Check for duplicate rows
duplicate_rows = df3[df3.duplicated()]
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")

# Drop duplicate rows
df3_no_duplicates = df3.drop_duplicates()
print(f"Shape of dataframe after removing duplicates: {df3_no_duplicates.shape}")

df3_no_duplicates
```

## HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python
import plotly.express as px

# for the year 2020
df_2020 = df3[df3["Year"] == 2020]

fig = px.scatter(
    df_2020,
    x="GDP per capita (current US$)",
    y="Life expectancy at birth, total (years)",
    hover_name="Country",
    title="Relationship between GDP per capita and Life Expectancy (2020)",
    labels={
        "GDP per capita (current US$)": "GDP per capita (US$)",
        "Life expectancy at birth, total (years)": "Life Expectancy (years)"
    },
    trendline="ols"
)

fig.show()
```

## Homework Question 3

Select a year of data. calculate total population of each region and make a pie chart to display the distribution of total population of all regions.

```python
import plotly.express as px

# for the year 2020
df_2020 = df3[df3["Year"] == 2020]

# Calculate total population for each region
df_population_by_region = df_2020.groupby("Region")["Population, total"].sum().reset_index()


fig = px.pie(
    df_population_by_region,
    values="Population, total",
    names="Region",
    title="Population Distribution by Region (2020)"
)

fig.show()
```
