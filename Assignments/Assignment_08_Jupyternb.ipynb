```python
import pandas as pd

# Load the dataset
df = pd.read_csv("fish.csv")
```

```python
# Display first few rows
df.head()


```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics for numerical data
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(x='Species', data=df)
plt.title('Fish Species Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Length1', y='Weight', hue='Species', data=df)
plt.title('Weight vs Length1')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
avg_width = df.groupby('Species')['Width'].mean().sort_values()
sns.barplot(x=avg_width.index, y=avg_width.values, palette='plasma')
plt.title('Average Width by Fish Species')
plt.xlabel('Species')
plt.ylabel('Average Width')
plt.xticks(rotation=45)
plt.show()
```
